```python
import concurrent.futures
import time


# Function to call your_function in parallel
def run_in_parallel(your_function, params):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        # Map the function to the parameters and execute in parallel
        results = executor.map(your_function, params)
        return list(results)  # Convert the results to a list
```

```python
import time
def generate_everything(questions):
  start_time = time.time()
  questions_hindi = run_in_parallel(get_hindi_translation, questions)
  answers = run_in_parallel(get_gpt4_response, questions)
  answers_hindi = run_in_parallel(get_hindi_translation, answers)
  end_time = time.time()
  print(f"Took {end_time - start_time} seconds to genereate everything for {len(questions)} questions.")
  return questions_hindi, answers_hindi
```
